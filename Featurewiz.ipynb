# Automatic Features Selection Using Featurewiz

### Github Link: https://github.com/AutoViML/featurewiz

```python
import pandas as pd
from featurewiz import featurewiz #Importing Featurewiz
```

## Data Loading

### Dataset Provided By rickstanley76 - https://github.com/rickystanley76

```python
data = pd.read_csv("./dataset/papermil_data_new.csv")
```

```python
data.head()
```

## Choosing The Best Features To Use For Machine Learning Using Featurewiz

```python
feature_names,train_data = featurewiz(data,'Y',corr_limit=0.7,verbose=2)
```

```python
feature_names
```

```python
print("Best Features To Choose:")
for i in feature_names:
    print(i)
```

```python
train_data
```

## Using Diabetes Dataset To Extract The Best Features For Regression

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes(as_frame=True)
```

```python
diabetes.frame.head()
```

```python
target = 'target'
```

```python
diabetes_features,diabetes_train = featurewiz(diabetes.frame,target,corr_limit=0.3,verbose=2)
```

```python
print("Best Features To Choose Of Diabates Datasets:")
for i in diabetes_features:
    print(i)
```
